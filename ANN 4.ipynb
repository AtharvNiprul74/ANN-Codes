**1.Email Spam Detection.**

```python
import numpy as np

emails = [
    "WIN a FREE prize now!",
    "Hello, let's meet at 5 PM.",
    "Claim your FREE PRIZE reward now!",
    "Project update: Please review the document.",
    "BIG WIN! Claim your lottery PRIZE today!"
]

def capital_count(email):
  count=0
  for i in email:
    if i.isupper():
      count+=1
  return count

def count_words(email):
  words = {"free", "win", "prize"}
  count = 0
  email_words = email.lower().split()
  for word in email_words:
    if word in words:
      count += 1
  return count


def len_email(email):
  return len(email)


email_features = np.array([[capital_count(email), count_words(email), len_email(email)] for email in emails])
mean = np.mean(email_features, axis=0)
std = np.std(email_features, axis=0)
std = np.where(std == 0, 1, std)

email_features=(email_features-mean)/std

x=email_features
y=np.array([1,0,1,0,1])

def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def perceptron():
  learn_rate=0.01
  epochs=1000
  features=x.shape[1]
  # print(x.shape[1])
  np.random.seed(40)
  # w=np.random.randn(features)
  # b=np.random.randn()
  w=np.zeros(features)
  b=0
  for epoch in range(epochs):
    for i in range(len(x)):
      weighted_sum=np.dot(w,x[i])+b

      y_pred=sigmoid(weighted_sum)
      # print(y_pred)

      error=y[i]-y_pred
      # print(error)

      w=w+learn_rate*(error*x[i])
      b=b+learn_rate*(error)

    # if epoch%100==0:
    #   print(f"Weight:{w},Bias:{b},error:{error}")

  return w,b

final_w,final_b=perceptron()
print(f"Final Weight:{final_w},Final Bias:{final_b}")

# Check output
# new_email="You Won A Bike as a Prize For FREE."
# new_email_features = np.array([[capital_count(new_email), count_words(new_email), len_email(new_email)]])
# norm_email_features=(new_email_features-mean)/std

# sum_val=np.dot(final_w,norm_email_features[0])+final_b

sample_email="You Won A Bike as a Prize For FREE."
sample_email_features = np.array([[capital_count(sample_email), count_words(sample_email), len_email(sample_email)]])
for i in range(len(sample_email_features)):
  sum=np.dot(final_w,x[i])+final_b

output=sigmoid(sum)
print(output)
if output > 0.5:
  print("Spam Email.")
else:
  print("Not Spam email.")

# **Calculate Accuracy**
# correct_predictions = 0
# for i in range(len(email_features)):
#     weighted_sum = np.dot(final_w, email_features[i]) + final_b
#     output = sigmoid(weighted_sum)

#     # Convert probability to class label
#     predicted_label = 1 if output >= 0.5 else 0

#     # Check if prediction is correct
#     if predicted_label == y[i]:
#         correct_predictions += 1

# # Compute Accuracy
# accuracy = (correct_predictions / len(y)) * 100
# print(f"Model Accuracy: {accuracy:.2f}%")
```

```python
import numpy as np

emails = [
    "WIN a FREE prize now!",
    "Hello, let's meet at 5 PM.",
    "Claim your FREE PRIZE reward now!",
    "Project update: Please review the document.",
    "BIG WIN! Claim your lottery PRIZE today!"
]

def capital_count(email):
  count=0
  for i in email:
    if i.isupper():
      count+=1
  return count

def count_words(email):
  words = {"free", "win", "prize"}
  count = 0
  email_words = email.lower().split()
  for word in email_words:
    if word in words:
      count += 1
  return count


def len_email(email):
  return len(email)

email_features = np.array([[capital_count(email), count_words(email), len_email(email)] for email in emails])

x=np.array(email_features)
y=np.array([1,0,1,0,1])

def step(x):
  return np.where(x>=0,1,0)

def perceptron():
  learn_rate=0.01
  epochs=1000
  features=x.shape[1]
  np.random.seed(40)
  # w=np.random.randn(features)
  # b=np.random.randn()
  w=np.zeros(features)
  b=0
  for epoch in range(epochs):
    for i in range(len(x)):
      weighted_sum=np.dot(w,x[i])+b

      y_pred=step(weighted_sum)
      # print(y_pred)

      error=y[i]-y_pred
      # print(error)

      w=w+learn_rate*(error*x[i])
      b=b+learn_rate*(error)

    # if epoch%100==0:
    #   print(f"Weight:{w},Bias:{b},error:{error}")
  return w,b

final_w,final_b=perceptron()
print(f"Final Weight:{final_w},Final Bias:{final_b}")

# Check output

sample_email="You Won A Bike as a Prize For FREE."
sample_email_features = np.array([[capital_count(sample_email), count_words(sample_email), len_email(sample_email)]])
for i in range(len(sample_email_features)):
  sum=np.dot(final_w,x[i])+final_b

output=step(sum)
print(output)
if output :
  print("Spam Email.")
else:
  print("Not Spam email.")
```

```python
import numpy as np

account_data = [
    {'credit_score': 700, 'annual_income': 50000, 'past_defaults': 0, 'loan_approved': 1},
    {'credit_score': 650, 'annual_income': 42000, 'past_defaults': 1, 'loan_approved': 0},
    {'credit_score': 720, 'annual_income': 60000, 'past_defaults': 0, 'loan_approved': 1},
    {'credit_score': 580, 'annual_income': 30000, 'past_defaults': 1, 'loan_approved': 0},
    {'credit_score': 750, 'annual_income': 75000, 'past_defaults': 0, 'loan_approved': 1},
]

account_features=np.array([[d['credit_score'],d['annual_income'],d['past_defaults']]for d in account_data])
y=np.array([d['loan_approved']for d in account_data])
print(y)

def normalize(column):
    column = np.array(column)
    mean = np.mean(column)
    std_dev = np.std(column)
    return (column - mean) / std_dev if std_dev != 0 else column



credit_scores=normalize([d['credit_score'] for d in account_data])
annual_incomes=normalize([d['annual_income'] for d in account_data])
past_defaults=[d['past_defaults'] for d in account_data]

account_features_normalized=np.column_stack((credit_scores,annual_incomes,past_defaults))


def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def perceptron():
  learn_rate=0.01
  epochs=1000
  features=account_features.shape[1]
  np.random.seed(40)
  # w=np.random.randn(features)
  # b=np.random.randn()
  w=np.zeros(features)
  b=0
  for epoch in range(epochs):
    for i in range(len(account_features)):
      weighted_sum=np.dot(w,account_features_normalized[i])+b

      y_pred=sigmoid(weighted_sum)
      # print(y_pred)

      error=y[i]-y_pred
      # print(error)

      w=w+learn_rate*(error*account_features_normalized[i])
      b=b+learn_rate*(error)

    # if epoch%100==0:
    #   print(f"Weight:{w},Bias:{b},error:{error}")

  return w,b

final_w,final_b=perceptron()
print(f"Final Weight:{final_w},Final Bias:{final_b}")

x=np.array([[710,55000,0]])
x_norm=normalize(x)
for i in range(len(x_norm)):
  sum=np.dot(final_w,x_norm[i])+final_b

output=sigmoid(sum)
print(output)
if output >= 0.5:
  print("Approved.")
else:
  print("Not Approved.")
```
